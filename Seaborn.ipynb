```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt
```

```python
#Scatter Plot
sns.scatterplot(x = 'total_bill',y = 'tip',data=tips)
plt.title('Scatter plot of Total bill vs Tip')
plt.show()
```

```python
#LinePLot
sns.lineplot(x='size',y = 'total_bill',data = tips)
plt.title('Line plot of total bill by size')
plt.show()
```

```python
#Categorial plots
sns.barplot(x='day',y='total_bill',data = tips)
plt.title('Bar plot of total bill by day')
plt.show()
```

```python
sns.barplot(x='size',y='total_bill',data = tips)
plt.title('Bar plot of total bill by day')
plt.show()
```

```python
#Boxplot
sns.boxplot(x='sex',y='total_bill',data=tips)
plt.title('Boxplot of total bill by sex')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Boxplot of total bill by day')
plt.show()
```

```python
#Violoin plot - it's basically tells the distribution of the data
sns.violinplot(x='day',y='total_bill',data = tips)
plt.title('Violin Plot of total bill by day')
plt.show()
```

```python
#Histogram
sns.histplot(tips['total_bill'],bins = 10,kde = True)
plt.title('Histogram of total bill ')
plt.show()
```

```python
#KDE plot
sns.kdeplot(tips['total_bill'],fill = True)
plt.show()
```

```python
#pair plot
#We use pair plot to show the relationship of variable to other different variables 
sns.pairplot(tips)
```

```python
#Heatmap
#it is only used when we have numerical data not for categorical data
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
#shows the correlation of size,tip and total_bill
```

```python
import pandas as pd
```

```python
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
plt.figure(figsize =(15,12))
sns.barplot(x='region',y = 'total_price',data = sales_df,estimator = sum)
plt.title('Total price by product')
plt.xlabel('region')
plt.ylabel('total_price')
plt.show()
           
```

```python
plt.figure(figsize = (10,6))
sns.boxplot(x='category',y = 'total_price',data = sales_df,color = 'purple')
plt.show()
```
